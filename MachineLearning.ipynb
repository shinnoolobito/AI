```python
data = pd.read_csv("/content/drive/My Drive/AI/car.data",header=None)
data.columns = ['buying','maint','doors','persons','lug_boot','safety','classes']
```

```python
data.buying.replace(('vhigh','high','med','low'),(1,2,3,4), inplace=True)
data.maint.replace(('vhigh','high','med','low'),(1,2,3,4), inplace=True)
data.doors.replace(('2','3','4','5more'),(1,2,3,4), inplace=True)
data.persons.replace(('2','4','more'),(1,2,3), inplace=True)
data.lug_boot.replace(('small','med','big'),(1,2,3), inplace=True)
data.safety.replace(('low','med','high'),(1,2,3), inplace=True)
```

```python
data.head()
```

```python
dataSet = data.values
np.random.shuffle(dataSet)
trainSet = dataSet[:1500]
testSet = dataSet[1500:]
```

```python
def calcDistancs(pointA, pointB, numOfFeature=6):
    tmp = 0
    for i in range(numOfFeature):
        tmp += (pointA[i] - pointB[i]) ** 2
    return math.sqrt(tmp)
```

```python
def kNearestNeighbor(trainSet, point, k):
    distances = []
    for item in trainSet:
        distances.append({
            "label": item[-1],
            "value": calcDistancs(item, point)
        })
    distances.sort(key=lambda x: x["value"])
    labels = [item["label"] for item in distances]
    return labels[:k]
```

```python
def findMostOccur(arr):
    labels = set(arr) # set label
    ans = ""
    maxOccur = 0
    for label in labels:
        num = arr.count(label)
        if num > maxOccur:
            maxOccur = num
            ans = label
    return ans    
```

```python
if __name__ == "__main__":
    trainSet,testSet
    numOfRightAnwser = 0
    for item in testSet:
        knn = kNearestNeighbor(trainSet, item, 6)
        answer = findMostOccur(knn)
        numOfRightAnwser += item[-1] == answer
        #print("label: {} -> predicted: {}".format(item[-1], answer))
    print("Accuracy", numOfRightAnwser/len(testSet))
```

K-NN dùng Scikit-learn

```python
X=data[data.columns[:-1]]
Y=data['classes']
X = X.to_numpy()
Y = Y.to_numpy()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6)
```

```python
knn.fit(x_train,y_train)
pred = knn.predict(x_test)
knn.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
avg_score=[]
for k in range(1,30):
    knn=KNeighborsClassifier(n_neighbors=k)
    score=cross_val_score(knn,x_train,y_train,scoring='accuracy')
    avg_score.append(score.mean())
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred)

print('Confusion matrix\n\n', cm)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plt.plot(range(1,30),avg_score)
plt.xlabel("n_neighbours")
plt.ylabel("accuracy")
```
